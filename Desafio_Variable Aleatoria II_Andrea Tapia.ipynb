### <span style="color:green">3</span>.- Considera los dos pares de variables anteriores y elimina en cada caso los faltantes y atípicos. Justifica en cada caso tu procedimiento.

#### <span style="color:blue">Se toman las variables Tasa de fertilidad y Tasa natalidad para hacer la comparación</span>



```python
# Anteriormente se eliminaron los valores nulos, se verifica la cantidad de valores nulos en cada columna seleecionada

df_columna = (df_limpio[['Fertility Rate', 'Birth Rate']].isnull().sum())
print(df_columna)
```

#### <span style="color:blue">A continuación se calcularán los rangos intercuantiles para excluir valores atipicos</span>


```python
# Se usa una función para 
'''Se define una función llamada 'intercuantil' para calcular los valores que están dentro del rango intercuartil
de una columna. Este rango se usa para identificar valores atípicos.'''

def intercuantil(columna):
    # Calcula el primer cuartil (Q1), que es el valor que deja el 25% de los datos por debajo de él.
    q25 = columna.quantile(0.25)
    
    # Calcula el tercer cuartil (Q3), que es el valor que deja el 75% de los datos por debajo de él.
    q75 = columna.quantile(0.75)
    
    # Calcula el rango intercuartil (IQR) restando Q1 de Q3. Este rango mide la dispersión en el 50% medio de los datos.
    rango = q75 - q25
    
    # Calcula el límite superior como Q3 + IQR. Cualquier valor por encima de este límite podría considerarse un valor atípico.
    limitesup = q75 + rango
    
    # Calcula el límite inferior como Q1 - IQR. Cualquier valor por debajo de este límite podría considerarse un valor atípico.
    limiteinf = q25 - rango
    
    # Crea una condición booleana que es True para los valores que están entre los límites inferior y superior.
    condicion = ((limiteinf <= columna) & (columna <= limitesup))
    
    # Devuelve la condición booleana que puede usarse para filtrar los datos.
    return condicion

'''Filtramos el DataFrame 'df_limpio' para mantener solo las filas donde los valores de 'Birth Rate' y 
'Fertility Rate'estén dentro de sus respectivos rangos intercuartiles. Se utiliza 'loc' para seleccionar 
las filas y columnas específicas.'''

df_compara = df_limpio.loc[intercuantil(df_limpio['Birth Rate']) & intercuantil(df_limpio['Fertility Rate']), 
                           ['Fertility Rate', 'Birth Rate','Life expectancy']]
df_compara
```

### <span style="color:green">4</span>.- Construye un gráfico para verificar la correlación entre estas variables, ahora con los datos preparados. Compara con lo obtenido anteriormente.

```python
# Gráfico de dispersión para verificar la correlación una vez excluidos los valores atípicos
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Birth Rate', y='Fertility Rate', hue='Life expectancy', data=df_compara, palette='viridis')
plt.title('Relación entre Tasa de Natalidad y Tasa de Fertilidad por Esperanza de Vida')
plt.show()
```

```python
# Gráfico de dispersión para verificar la correlación previo a la exclusión de valores atípicos
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Birth Rate', y='Fertility Rate', hue='Life expectancy', data=df_limpio, palette='viridis')
plt.title('Relación entre Tasa de Natalidad y Tasa de Fertilidad por Esperanza de Vida')
plt.show()
```

#### <span style="color:purple">Comparación:</span>

- En el gráfico que muestra la relación entre la Tasa de Natalidad y la Tasa de Fertilidad antes de excluir los valores atípicos se pueden observar algunos puntos dispersos fuera de la tendencia principal, lo que indica la presencia de valores atípicos. La dispersión de los datos es un poco más amplia, especialmente en las tasas más altas.

- En el grafico que muestra la relación entre la Tasa de Natalidad y la Tasa de Fertilidad después de excluir los valores atípicos, al eliminar los valores atípicos, la dispersión de los puntos se reduce, lo que sugiere una relación más clara.

- A pesar de visualizarse esta diferencia, es muy pequeña por lo que no se podría decir que es una diferencia significativa.

#### <span style="color:blue">Preparación de datos</span>

```python
# Verificar el tipo de dato de las dos columnas a utilizar
tipos_de_datos = df_filtrado[['GDP', 'Population']].dtypes
print(tipos_de_datos)
```

```python
# Eliminamos los símbolos de moneda y comas en la columna 'GDP'
df_filtrado['GDP'] = df_filtrado['GDP'].str.replace('$', '').str.replace(',', '').str.strip()

# Convertir las columnas 'GDP' y 'Population' a tipo float
df_filtrado['GDP'] = df_filtrado['GDP'].astype(float)
df_filtrado['Population'] = df_filtrado['Population'].str.replace(',', '').astype(float) # Reemplazamos las comas de esta columna 

# Verificar los tipos de datos después de la conversión
tipos_de_datos = df_filtrado[['GDP', 'Population']].dtypes  
print(tipos_de_datos)

```

```python
plt.figure(figsize=(20, 15))
sns.scatterplot(x='GDP', y='Population', hue='CPI', data=df_compara2, palette='deep')
plt.title('Relación entre Población Total y PIB por CPI')
plt.show()
```

- Primero, se prepararon los datos verificando su tipo y luego se convirtieron a tipo numérico para facilitar los cálculos posteriores.

- A continuación, se utilizó la función *intercuantil* para identificar y excluir los valores atípicos, lo que nos ayuda a obtener una representación más precisa y confiable de los datos.

- Finalmente, se graficaron los datos filtrados utilizando un gráfico de dispersión. Se escoge este tipo de gráfico porque es ideal para este tipo de análisis, ya que entrega una forma clara y directa de visualizar la relación entre la Población Total y el PIB de los países. El gráfico de dispersión permite identificar fácilmente patrones, tendencias, y posibles valores atípicos, lo que ayuda a entender mejor la relación entre estas dos variables.
